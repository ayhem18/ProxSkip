```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import pickle


# Ref: https://github.com/gsmalinovsky/ProxSkip-Public/blob/main/ProxSkip.ipynb
sns.set(
    style="whitegrid", 
    context="talk", 
    font_scale=1.2, 
    palette=sns.color_palette("bright"), 
    color_codes=False
)

matplotlib.rcParams['font.family'] = 'sans-serif'
matplotlib.rcParams['font.sans-serif'] = 'DejaVu Sans'
matplotlib.rcParams['mathtext.fontset'] = 'cm'
matplotlib.rcParams['lines.markersize'] = 10
```

```python
prox1 = pickle.load(open('resnet18/assets/resnet18.losses.p500.pt', 'rb'))
prox2 =  pickle.load(open('resnet18/assets/resnet18.losses.p50.pt', 'rb'))
prox3 = pickle.load(open('resnet18/assets/resnet18.losses.p10.pt', 'rb'))
prox4 = pickle.load(open('resnet18/assets/resnet18.losses.p10000.pt', 'rb'))
```

```python
fig = plt.figure(figsize=(10, 10))

sns.lineplot(prox1, marker='*', label=r"$p=\frac{1}{500}$")
sns.lineplot(prox2, marker='*', label=r"$p=\frac{1}{50}$")
sns.lineplot(prox3, marker='*', label=r"$p=\frac{1}{10}$")
sns.lineplot(prox4, marker='*', label=r"$p=\frac{1}{10000}$")

plt.title("Training ResNet-18 on CIFAR-10 with Scaffnew", pad=20)
plt.xlabel("Communication Rounds", labelpad=20)
plt.ylabel(r"$\nabla f(x)$", labelpad=20)
plt.tight_layout()
plt.show()
```
